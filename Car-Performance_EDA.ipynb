# barplot

```python
# barplot represents how a numerical variable relates to a categorical variable.
# thus, we have to put values for both x and y axis whereas in countplots, y axis is default count
# it is a bivariate analysis whereas count plot is univariate analysis
# y axis is numerical variable where as x axis is categorical variable
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="origin",data=df,y="mpg")

#the black vertical line above the bars represents the minimum and maximum values of mpg (y axis) in particular nations 
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="weight",data=df,y="displacement")

#example of passing numerical variable on x axis ---> kinda useless cuz clumsy representation
```

# Histogram

Students pass a test if they score 50% or more.The marks of a large no of students 
were sampled and the mean and sd were calculated 42 % and 8% respectively.
assume the data is normally distributed.
what percentage of students pass the test?

- Answer = 16%

The mean june midday temp at your hometown is 36 deg c.and the SD is 3 deg c. 
assume the data is normally distributed . How many days in june 
would you expect the temp between 39 and 42 deg C.?

- Answer = about 13% of the days that is about 4 days

```python
plt.figure(figsize=(10,4))
sns.histplot(data=df,x="mpg",kde=True,bins=10)  #kde = true means the line will show on the graph, and bins mean number of
                                                                                                          #bars
plt.title("Distribution of MPG")

#this is right skewed data
```

```python
plt.figure(figsize=(10,4))
sns.histplot(data=df,x="mpg",kde=True,bins=10,hue="origin")
plt.title("Distribution of MPG")
```

```python
#by default y is count but we can change it to probability by using the stat parameter

plt.figure(figsize=(10,4))
sns.histplot(data=df,x="mpg",kde=True,bins=10,hue="origin",stat="probability")
plt.title("Distribution of MPG")
```
